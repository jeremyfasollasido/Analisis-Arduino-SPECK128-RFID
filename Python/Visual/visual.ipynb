```python
import pandas as pd
import matplotlib.pyplot as plt
import re

file_path = "../../Data_CSV_128/avg_csv_128/avg_csv_128.csv"
data = pd.read_csv(file_path)

data['Byte'] = data['file_name'].apply(lambda x: int(re.search(r'_(\d+)', x).group(1)) if pd.notnull(x) else 0)

data.drop(columns=['file_name'], inplace=True)

data.set_index('Byte', inplace=True)

for column in data.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data[column], marker='o')
    plt.title(f'{column} vs Byte')
    plt.xlabel('Byte')
    plt.ylabel(column)
    plt.grid(True)
    plt.show()
```
